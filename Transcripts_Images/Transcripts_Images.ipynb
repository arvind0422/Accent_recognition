```python
import numpy as np
import os
```

```python
for num in range(1,61):
    lines = [f.rstrip(". \n")+" " for f in open("./uniq_sent_60spkrs/x"+"{0:02}".format(num), "r").readlines()]
    os.system("mkdir spkr"+str(num))
    f = open("spkr"+str(num)+"/x"+"{0:03}".format(num)+".txt", "w")
    for line in lines:
        spaces = np.array([pos for pos, char in enumerate(line) if char == " "][0:]) # Find position of spaces.
        num_lines = int(max(spaces)/70)+1 # Estimate number of lines.
        breaks=[]
        for i in range(num_lines):
            breaks.append(max((spaces-70*(i+1))[(spaces-70*(i+1)) < 0])+70*(i+1)) # Estimate break positions.
        num_zeros = 7-num_lines
        front = int(num_zeros/2) # Empty lines before.
        back = num_zeros - front # Empty lines after.
        for i in range(front):
            f.write("\n")
        prev_break = 0
        for i in range(len(breaks)):
            print(line[prev_break:breaks[i]])
            f.write(line[prev_break:breaks[i]])
            f.write("\n")
            prev_break = breaks[i]+1
        for i in range(back):
            f.write("\n")
    f.close()
```
